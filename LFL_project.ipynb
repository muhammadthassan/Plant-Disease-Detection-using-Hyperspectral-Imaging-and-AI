```python
# Normalize the intensity values to the range [0, 1].
nir_image_normalized = (nir_data - nir_data.min()) / (nir_data.max() - nir_data.min())

# Display the normalized NIR image using matplotlib.
plt.imshow(nir_image_normalized.squeeze())
plt.title('Normalized NIR Image')
plt.colorbar()
plt.show()
```

```python

# Normalize the intensity values to the range [0, 1].
red_image_normalized = (red_data - red_data.min()) / (red_data.max() - red_data.min())

# Display the normalized RED image using matplotlib.
plt.imshow(red_image_normalized.squeeze())
plt.title('Normalized RED Image')
plt.colorbar()
plt.show()
```

```python
# histogram for the NIR data
# Get the flattened NIR data.
flattened_nir_data = nir_image_normalized.values.flatten()

# Plot the histogram of the flattened NIR data.
plt.hist(flattened_nir_data)
plt.xlabel('NIR Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized NIR Data')
plt.show()


# histogram for the RED data
# Get the flattened RED data.
flattened_red_data = red_image_normalized.values.flatten()

# Plot the histogram of the flattened NIR data.
plt.hist(flattened_red_data)
plt.xlabel('RED Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized RED Data')
plt.show()
```

```python
R = red_image_normalized.squeeze()
NIR = nir_image_normalized.squeeze()

print(f"Shape of Red image: {R.shape}")
print(f"Shape of NIR image: {NIR.shape}")
```

```python
# Crop the Image vertically by Selecting the targeted field from the original image

# Define the desired crop range in the x-axis
x_start = 200
x_end = 900

# Perform the crop operation
cropped_image_nir = NIR[:, x_start:x_end]
cropped_image_r = R[:, x_start:x_end]

# Check the new shape of the cropped image
print("Cropped image shape:", cropped_image_r.shape)

#ploting Red Image
plt.imshow(cropped_image_r)
plt.colorbar()
plt.title("RED Image")
plt.show()

#ploting Red Image
plt.imshow(cropped_image_nir)
plt.colorbar()
plt.title("NIR Image")
plt.show()
```

```python
# Calculate the NDVI vegetation index.
ndvi = (cropped_image_nir - cropped_image_r) / (cropped_image_nir + cropped_image_r)

print(f"Shape of NDVI image: {ndvi.shape}")
print(f"NDVI values range from {ndvi.min()} to {ndvi.max()}.")
print(f"NDVI values range from {ndvi.min()} to {ndvi.max()}.")
print(ndvi)
```

```python
#ploting NDVI 
plt.imshow(ndvi.squeeze())
plt.colorbar()
plt.title("NDVI Image")
plt.show()
```

```python
ndvi.shape
```

```python
flattened_ndvi= ndvi.values.flatten()

plt.hist(flattened_ndvi, bins=50, color='green')
plt.xlabel('Vegetation Index')
plt.ylabel('Frequency')
plt.title('Histogram of Vegetation Index')
plt.show()
```

```python
# Calculate the mean and standard deviation of the flattened image.
mean = np.mean(flattened_ndvi)
std = np.std(flattened_ndvi)

# Print the mean and standard deviation of the flattened image.
print("Mean:", mean)
print("Standard deviation:", std)
```

```python
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu

# Set a threshold for the NDVI value. Any pixel with an NDVI value below the threshold will be considered soil.

# Mask the image with the thresholded NDVI value. This will create a binary image where pixels with an NDVI value below 
# the threshold are set to 0 (black) and pixels with an NDVI value above the threshold are set to 1 (white).
# Create a threshold for the NDVI vegetation index.
ndvi_data = np.nan_to_num(ndvi)


ndvi_data = ndvi_data.squeeze()
print(ndvi_data.shape)

threshold = threshold_otsu(ndvi_data)
print(threshold)
```

```python
threshold = -0.08203125

# Mask the image with the thresholded NDVI value.
masked_image = ndvi_data <= threshold

# Remove the soil from the image by setting all pixels with a value of 0 to NaN.
soil_free_image = ndvi_data.copy()
soil_free_image[masked_image] = np.nan


#ploting soil_free_image 
plt.imshow(soil_free_image)
plt.colorbar()
plt.title("soil_free Image")
plt.show()
```
